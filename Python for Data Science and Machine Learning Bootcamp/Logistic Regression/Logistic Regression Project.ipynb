# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
##I will consider the hour intervals as categorical data
#so 00:00:00-0:59:59 =>0 ; 1:00:00-1:59:59 => 1 etc.
```

```python
hour_intervals = pd.to_datetime(ad_data['Timestamp']).apply(lambda x: x.hour)
ad_data.drop('Timestamp', inplace=True, axis=1)
#this way we took the hour from the timestamp
```

```python
ad_data = pd.concat([ad_data,hour_intervals], axis=1)
ad_data.head()
```

```python
ad_data.head(n=1)
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income',
       'Daily Internet Usage', 'Male',
       'Timestamp']]
y = ad_data['Clicked on Ad']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

** Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression()
```

```python
logr.fit(X_train, y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
predictions = logr.predict(X_test)
```

** Create a classification report for the model.**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
